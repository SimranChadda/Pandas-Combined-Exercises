### <font color="purple">**Question3: Group data using custom function:**</font>
Here, the requirement is to create four groups based on **"imdb_rating"**

- 1 <= imdb_rating <= 3.9 ---> 'Poor'
- 4 <= imdb_rating <= 7.9 ---> 'Average'
- 8 <= imdb_rating <= 10 --->  'Good'
- If none of above satisfy, then return 'others'

```python
def condition(imdb_rating):
    if(1<= imdb_rating<=3.9):
        return "==> Poor"
    elif(4 <= imdb_rating <= 7.9 ):
        return "==> Average"
    elif(8 <= imdb_rating <= 10):
        return "==>Good"
    else:
        return "others"
```

```python
df1['imdb_category'] = df1['imdb_rating'].apply(condition)
```

```python
df1
```

```python
g1= df1.groupby('imdb_category')
```

```python
g1.size()
```

```python
df1[df1['imdb_category'] == '==> Average']
```

```python
df1[df1['imdb_category'] == '==>Good']
```

```python
df1[df1['imdb_category'] == '==> Poor']
```

```python
df1[df1['imdb_category'] == 'others']
```
